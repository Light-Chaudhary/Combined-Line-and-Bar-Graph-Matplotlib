```python
# Horizontal Bar Graph

import matplotlib.pyplot as plt
import numpy as np
years = [2015, 2016, 2017, 2018, 2019, 2020]

kohli_runs = [850, 900, 1000, 1100, 950, 1050]
rohit_runs = [700, 850, 900, 950, 1000, 1100]
dhawan_runs = [650, 700, 750, 800, 850, 900]

y = np.arange(len(years))
height = 0.25

plt.barh(y - height, kohli_runs, height = height, label = "Kohli Runs")
plt.barh(y, rohit_runs, height = height,  label = "Rohit Runs")
plt.barh(y + height, dhawan_runs, height = height, label = "Dhawan Runs")
plt.yticks(y, years)
plt.xlabel("Runs")
plt.ylabel("Year")
plt.legend()
plt.title("Run comparison")
plt.grid(True)
plt.tight_layout()
plt.show()
```
